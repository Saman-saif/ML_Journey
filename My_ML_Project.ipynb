```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('pakistan_trafic_challan.csv')
df_small = df.sample(200, random_state=42)

# Boxplot banao
plt.figure(figsize=(7, 4))
plt.boxplot(df_small['Fine_Amount_PKR'])
plt.title('Fine Amount - Outliers Detection')
plt.ylabel('Fine Amount (PKR)')
plt.show()
# IQR = Interquartile Range
Q1 = df_small['Fine_Amount_PKR'].quantile(0.25)
Q3 = df_small['Fine_Amount_PKR'].quantile(0.75)
IQR = Q3 - Q1

# Outlier boundaries
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

print(f"Q1: {Q1}")
print(f"Q3: {Q3}")
print(f"IQR: {IQR}")
print(f"Lower Boundary: {lower}")
print(f"Upper Boundary: {upper}")

# Outliers 
outliers = df_small[
    (df_small['Fine_Amount_PKR'] < lower) | 
    (df_small['Fine_Amount_PKR'] > upper)
]
print(f"\nTotal Outliers: {len(outliers)}")
print(outliers['Fine_Amount_PKR'])
df_small['Fine_Amount_PKR'] = df_small['Fine_Amount_PKR'].clip(lower=lower,upper= upper)
print("maximum value bad mn", df_small['Fine_Amount_PKR'].max())
plt.figure(figsize= (5,5))
plt.boxplot(df_small['Fine_Amount_PKR'])
plt.title('Total Fine Amount Outliers Handled')
plt.ylabel('Total Fine Amount')
plt.show()
```

```python
import pandas as pd 
import matplotlib.pyplot as plt
from scipy import stats
df = pd.read_csv('pakistan_trafic_challan.csv')
df_small = df.sample(200,random_state =42)
plt.figure(figsize=(8,5))
plt.boxplot(df_small['Amount_Collected_PKR'])
plt.title('Total Fine Amount Calculated')
plt.ylabel('Total Amount Calculted(PKR)')
plt.show()
Q1 = df_small['Amount_Collected_PKR'].quantile(0.25)
Q3 = df_small['Amount_Collected_PKR'].quantile(0.75)
IQR = Q3 - Q1

print(f"Q1: {Q1}")
print(f"Q3: {Q3}")
print(f"IQR: {IQR}")
lower = Q1 - (1.5 * IQR)
upper = Q3 + (1.5 * IQR)

print(f"Lower Boundary: {lower}")
print(f"Upper Boundary: {upper}")
df_small['Amount_Collected_PKR'] = df_small['Amount_Collected_PKR'].clip(
    lower=lower,
    upper=upper
)

print("Max value baad mein:", 
      df_small['Amount_Collected_PKR'].max())
plt.figure(figsize=(6,4))
plt.boxplot(df_small['Amount_Collected_PKR'])
plt.title('After caping Amount_Calculted')
plt.ylabel('Amount_Calculated(PKR)')
plt.show()
df_small['Amount_Collected_PKR'].hist()
plt.show()
mean = df_small['Amount_Collected_PKR'].mean()
median = df_small['Amount_Collected_PKR'].median()
skewness = df_small['Amount_Collected_PKR'].skew()
print(f"mean:  {mean}")
print(f"median:   {median}")
print(f"skewness:   {skewness}")
z_scores = stats.zscore(df_small['Amount_Collected_PKR'])
outliers_z = df_small[abs(z_scores)>3]
print(f"total outliers :  {len(outliers_z)}")
print(outliers_z['Amount_Collected_PKR'])


```

```python
from scipy import stats
import pandas as pd

df = pd.read_csv('pakistan_trafic_challan.csv')
df_small = df.sample(200, random_state=42)

# Z-Score calculate karo
z_scores = stats.zscore(df_small['Fine_Amount_PKR'])

# Z-Score 3 se zyada = Outlier
outliers_z = df_small[abs(z_scores) > 3]

print(f"Total Outliers: {len(outliers_z)}")
print(outliers_z['Fine_Amount_PKR'])
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
df = pd.read_csv('pakistan_trafic_challan.csv')
df_small = df.sample(1000, random_state=42)
scaler = MinMaxScaler()
df_small[['Fine_Amount_PKR', 'Amount_Collected_PKR']] = scaler.fit_transform(
    df_small[['Fine_Amount_PKR', 'Amount_Collected_PKR']]
)

print(df_small[['Fine_Amount_PKR', 'Amount_Collected_PKR']].head(10))

# Standardization Practice
scaler = StandardScaler()
df_small[['Fine_Amount_PKR', 'Amount_Collected_PKR']] = scaler.fit_transform(
  df_small[['Fine_Amount_PKR', 'Amount_Collected_PKR']]  
)
print('After Standardization output ' ,df_small[['Fine_Amount_PKR', 'Amount_Collected_PKR']].head(10))
```
